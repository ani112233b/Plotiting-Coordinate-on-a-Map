```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
m = Basemap(width=1900000,height=1900000,projection='lcc',
            resolution='f',lat_1=12,lat_2=18,lat_0=15,lon_0=73)
m.shadedrelief(alpha = 0.9)
m.drawparallels(range(-90, 90, 5), labels=[1, 0, 0, 0], linewidth=0.5, color='gray')
m.drawmeridians(range(-180, 180, 5), labels=[0, 0, 0, 1], linewidth=0.5, color='gray')

# Map (long, lat) to (x, y) for plotting
x1, y1 = m(73.79358333,15.42136111)
plt.plot(x1, y1, marker = 'o', markersize=15)
plt.text(x1, y1, '    Dona Paula Bay', color = 'black', fontsize=10)
plt.savefig("DonaPaula.png", format="png", dpi=1800)
```

```python
#m = plt.figure(figsize=(5, 5), dpi = 1200)
m = Basemap(width=1900000,height=1900000,projection='lcc',
            resolution='f',lat_1=12,lat_2=18,lat_0=15,lon_0=73)
# Map (long, lat) to (x, y) for plotting
x1, y1 = m(73.79358333,15.42136111)
#x2,y2 = m(94.55697, 26.81446)
plt.plot(x1, y1, marker = 'o', markersize=15)
plt.text(x1, y1, '    Dona Paula Bay', color = 'black', fontsize=10)
m.drawcoastlines()
# draw a boundary around the map, fill the background.
# this background will end up being the ocean color, since
# the continents will be drawn on top.
m.drawmapboundary(fill_color='aqua')
# fill continents, set lake color same as ocean color.
m.fillcontinents(color='coral',lake_color='aqua')
plt.savefig("DonaPaula2.png", format="png", dpi=1800)
```
